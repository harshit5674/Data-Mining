```python
import numpy as np
import pandas as pd
from sklearn import model_selection
```

```python
dp=pd.read_csv(r'Cars.csv')
print(dp)
```

```python
cla=[]
cla=dp.clazz.unique()
```

```python
inp=np.asarray(dp.iloc[:,0:6]).astype('str')
oup=np.asarray(dp.iloc[:,6]).astype('str')
print(inp)
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(inp, oup,test_size=0.3)
```

```python
dicti=[dict() for x in range(6)]
names=[]
for x in dp:
    names.append(str(x))
    print(x)
names.pop()
```

```python
count=0
for x in names:
    na=[]
    na=dp[x].unique()
    for y in na:
        dicti[count][y]=[0,0,0,0]
    count=count+1
```

```python
m=X_train.shape[0]
tr={
    'unacc':0,
    'acc':1,
    'vgood':2,
    'good':3
}
tri={
    0:'unacc',
    1:'acc',
    2:'vgood',
    3:'good'
}
for x in range(6):
    for y in range(m):
        r=y_train[y]
        r=tr[r]
        dicti[x][str(X_train[y,x])][r]=dicti[x][str(X_train[y,x])][r]+1
```

```python
count=0
for x in range(m):
    if X_train[x,0]=='vhigh' and y_train[x]=='unacc':
        count=count+1
print(count)
```

```python
print(dicti[0]['vhigh'][0])
```

```python
count=0
tr=dict()
cu=0
for x in cla:
    count=0
    for y in range(m):
        if y_train[y]==x:
            count=count+1
    tr[cu]=count
    cu=cu+1
```

```python
pred=0
tot_pred=y_test.shape[0]
print(tot_pred)
```

```python
print(tr)
```

```python
for i in range(4):
    tr[i]=tr[i]/(m*1.0)
```

```python
print(tr)
```

```python
print(count)
print(pred)
print(tot_pred)
ans=pred/(1.0*tot_pred)
print("Accuracy is "+str(ans))
```

```python
count=0
trpro={
    0:tr[0],
    1:tr[1],
    2:tr[2],
    3:tr[3]
}
for i in range(tot_pred):
    for j in range(4):
        for k in range(6):
            trpro[j]=(trpro[j]*1.0*dicti[k][X_test[i,k]][j])/(1.0*tr[j])
    best=0.0
    prediction=0
    for r in range(4):
        best=max(best,trpro[r])
        if(best==trpro[r]):
            prediction=r
    if(tri[prediction]==y_test[i]):
        pred=pred+1
    count=count+1
    trpro={
    0:tr[0],
    1:tr[1],
    2:tr[2],
    3:tr[3]
    }
print(count)
```

```python
print(count)
print(pred)
print(tot_pred)
ans=pred/(1.0*tot_pred)
print("Accuracy is "+str(100*ans))
```
